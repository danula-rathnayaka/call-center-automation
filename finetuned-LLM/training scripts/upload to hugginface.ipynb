```python
# Run this in Google Colab after Notebook 1 completes
from huggingface_hub import HfApi, login
import os

HF_TOKEN    = 
HF_USERNAME = 'Winindux'
login(token=HF_TOKEN)

api     = HfApi()
REPO_ID = f'{HF_USERNAME}/emowoz-llama2-data'

# Create a private dataset repo on HuggingFace
api.create_repo(
    repo_id=REPO_ID,
    repo_type='dataset',
    private=True,        # private so only you can access it
    exist_ok=True
)

# Upload every file from Google Drive to HF Hub
BASE_DIR = '/content/drive/MyDrive/emowoz_llama2'

files_to_upload = [
    ('data/train/sft_train.jsonl',       'data/train/sft_train.jsonl'),
    ('data/val/sft_val.jsonl',           'data/val/sft_val.jsonl'),
    ('data/test/sft_test.jsonl',         'data/test/sft_test.jsonl'),
    ('data/dpo_pairs/dpo_pairs.jsonl',   'data/dpo_pairs/dpo_pairs.jsonl'),
    ('prompt_utils.py',                  'prompt_utils.py'),
    ('data_summary.json',                'data_summary.json'),
]

for local_rel, hf_path in files_to_upload:
    local_full = os.path.join(BASE_DIR, local_rel)
    if os.path.exists(local_full):
        api.upload_file(
            path_or_fileobj=local_full,
            path_in_repo=hf_path,
            repo_id=REPO_ID,
            repo_type='dataset',
        )
        print(f'Uploaded: {hf_path}')
    else:
        print(f'MISSING (skipped): {local_full}')

print(f'\nAll done. View at: https://huggingface.co/datasets/{REPO_ID}')
```
